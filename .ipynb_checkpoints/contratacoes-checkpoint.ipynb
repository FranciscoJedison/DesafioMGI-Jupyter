```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
import warnings
import contextlib
import io

# Ignorar warnings do Python
warnings.filterwarnings("ignore")

# Redirecionar mensagens de stderr (onde o Seaborn imprime o FutureWarning)
f = io.StringIO()
with contextlib.redirect_stderr(f):
    import seaborn as sns
    import matplotlib.pyplot as plt


# Configurações visuais globais
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Função auxiliar para adicionar rótulos nas barras
def add_bar_labels_horizontal(ax, fmt='{:,.0f}'):
    for p in ax.patches:
        width = p.get_width()
        ax.text(width + (0.01*width), p.get_y() + p.get_height()/2,
                fmt.format(width).replace(",", "."), va='center')
```

```python
with open("contratacoes.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.json_normalize(data['resultado'])

# Tratamento dos dados
df['dataAberturaPropostaPncp'] = pd.to_datetime(df['dataAberturaPropostaPncp'], errors="coerce")
df['dataEncerramentoPropostaPncp'] = pd.to_datetime(df['dataEncerramentoPropostaPncp'], errors="coerce")
df = df.dropna(subset=['valorTotalEstimado', 'valorTotalHomologado'])
df['valorTotalEstimado'] = pd.to_numeric(df['valorTotalEstimado'], errors="coerce")
df['valorTotalHomologado'] = pd.to_numeric(df['valorTotalHomologado'], errors="coerce")
df = df[(df['valorTotalEstimado'] > 0) & (df['valorTotalHomologado'] > 0)]
df = df[(df['valorTotalEstimado'] < 1e7) & (df['valorTotalHomologado'] < 1e7)]

df['diferenca_valor'] = df['valorTotalEstimado'] - df['valorTotalHomologado']
df['duracao_dias'] = (df['dataEncerramentoPropostaPncp'] - df['dataAberturaPropostaPncp']).dt.days
df['ano_mes'] = df['dataAberturaPropostaPncp'].dt.to_period('M')
df['mes_abertura'] = df['dataAberturaPropostaPncp'].dt.month
```

# __1.1.Distribuição de Contratações por Modalidade__
Este gráfico mostra a quantidade de contratações para cada modalidade.


```python
modalidade_counts = df['modalidadeNome'].value_counts()
ax = sns.barplot(x=modalidade_counts.values, y=modalidade_counts.index, palette="coolwarm", hue=None, legend=False)
ax.set_title("Distribuição de Contratações por Modalidade")
ax.set_xlabel("Quantidade de Contratações")
ax.set_ylabel("Modalidade")
for i, v in enumerate(modalidade_counts.values):
    ax.text(v + (0.01*v), i, f'{v}', va='center')
plt.show()
```

# __1.2.Comparação entre Valores Estimados e Homologados__
Este gráfico de linha mostra a evolução dos valores estimados e homologados ao longo do tempo.


```python
valores_tempo = df.groupby('ano_mes')[['valorTotalEstimado', 'valorTotalHomologado']].sum()
ax = valores_tempo.plot(kind="line", marker="o")
ax.set_title("Comparação: Valor Estimado vs Homologado ao longo do tempo")
ax.set_xlabel("Ano-Mês")
ax.set_ylabel("Valores (R$)")
ax.legend(["Estimado", "Homologado"])
ax.tick_params(axis="x", rotation=45)
ax.get_yaxis().set_major_formatter(
    mtick.FuncFormatter(lambda x, _: f'R$ {x:,.0f}'.replace(",", "."))
)
plt.show()
```

O gráfico compara os valores planejados (valorTotalEstimado) e efetivamente homologados (valorTotalHomologado) ao longo do tempo (ano-mês). O gráfico de linha exibe ambos os valores com rótulos posicionados para evitar sobreposição. 
Permite identificar discrepâncias sistemáticas entre estimativa e homologação, indicando possíveis ajustes ou revisão de orçamentos futuros.

# __1.4.Valores Homologados por Amparo Legal__
Este gráfico apresenta o valor total homologado de acordo com o amparo legal.


```python
amparo = df.groupby("amparoLegalNome")['valorTotalHomologado'].sum().sort_values(ascending=False).reset_index()
ax = sns.barplot(data=amparo, x="valorTotalHomologado", y="amparoLegalNome", palette="tab20", hue=None, legend=False)
ax.set_title("Valores Homologados por Amparo Legal")
ax.set_xlabel("Valor Total Homologado (R$)")
ax.set_ylabel("Amparo Legal")
formatter = mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'.replace(",", "."))
ax.xaxis.set_major_formatter(formatter)
add_bar_labels_horizontal(ax)
plt.show()
```

```python
# =============================
# 1. Importar bibliotecas
# =============================
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Configurações visuais
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
pd.set_option('display.float_format', '{:,.2f}'.format)  # evita notação científica
```

```python
# =============================
# 2. Carregar dados
# =============================
with open("ItensContratacoes1.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.json_normalize(data['resultado'])

# Preenchendo nulos para evitar erros
df['valorUnitarioEstimado'] = df['valorUnitarioEstimado'].fillna(0)
df['valorTotal'] = df['valorTotal'].fillna(0)
df['temResultado'] = df['temResultado'].fillna('Sem Informação')
df['orcamentoSigiloso'] = df['orcamentoSigiloso'].fillna(False)
df['materialOuServicoNome'] = df['materialOuServicoNome'].fillna("Não informado")

# Converter datas
df['dataInclusaoPncp'] = pd.to_datetime(df['dataInclusaoPncp'])
df['mes'] = df['dataInclusaoPncp'].dt.to_period("M").dt.to_timestamp()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# =============================
# 1. Ler o arquivo JSON
# =============================
with open("resultadositenscontratacoes1.json", "r", encoding="utf-8") as f:
    data = json.load(f)

if isinstance(data, dict) and "resultado" in data:
    df = pd.DataFrame(data["resultado"])
else:
    df = pd.DataFrame(data)

```

```python
# =============================
# 2. Tratamento inicial
# =============================
num_cols = ["valorUnitarioHomologado", "valorTotalHomologado"]
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0)

if "dataResultadoPncp" in df.columns:
    df["dataResultadoPncp"] = pd.to_datetime(df["dataResultadoPncp"], errors="coerce")
    df["ano_mes"] = df["dataResultadoPncp"].dt.to_period("M").astype(str)

# Função para formatar valores monetários
def formatar_moeda(x, _):
    return f"R${x:,.0f}".replace(",", ".")
```

```python
# =============================
# 3. Funções auxiliares
# =============================
def add_values_bars(ax, fmt="R${:,.0f}"):
    for p in ax.patches:
        ax.annotate(fmt.format(p.get_width()),
                    (p.get_width(), p.get_y() + p.get_height() / 2),
                    ha='left', va='center', fontsize=10, xytext=(5, 0), textcoords='offset points')

def add_values_line(ax, x, y, fmt="R${:,.0f}"):
    for xi, yi in zip(x, y):
        ax.text(xi, yi, fmt.format(yi), ha='center', va='bottom', fontsize=9)
```

# __Análise — Endpoints 3 e 3.1 Resultados de Itens Contratações__
# <p align="justify">Os dados trazem resultados dos itens contratados (fornecedores vencedores, valores homologados, natureza jurídica, porte dos fornecedores, origem dos produtos/serviços, descontos aplicados, entre outros).</p>

# __3.1.Principais Fornecedores__
Gráfico mostrando os principais fornecedores por valor homologado.

```python
# --- Fornecedores ---
if "nomeRazaoSocialFornecedor" in df.columns:
    fornecedores_valor = df.groupby("nomeRazaoSocialFornecedor")["valorTotalHomologado"] \
                          .sum().sort_values(ascending=False)

    plt.figure(figsize=(10, 6))
    colors = sns.color_palette("tab10", n_colors=10)
    ax = fornecedores_valor.head(10).plot(kind="barh", color=colors)
    ax.xaxis.set_major_formatter(mtick.FuncFormatter(formatar_moeda))
    plt.title("10 Principais Fornecedores por Valor Homologado")
    plt.xlabel("Valor Total Homologado (R$)")
    plt.ylabel("Fornecedor")
    plt.gca().invert_yaxis()
    add_values_bars(ax)
    plt.tight_layout()
    plt.subplots_adjust(left=0.3) 
    plt.show()
```

Agrupamento por nomeRazaoSocialFornecedor e soma de valorTotalHomologado. Exibição dos 10 maiores em gráfico de barras horizontais. Identifica fornecedores que receberam os maiores repasses por contratos (maior participação em termos financeiros).

---

# __3.2.Valores Homologados por Natureza Jurídica__
Gráfico mostrando ordenado mostrando gasto por tipo de entidade.

```python
# --- Valores por Natureza Jurídica ---
if "naturezaJuridicaNome" in df.columns:
    valores_natureza = df.groupby("naturezaJuridicaNome")["valorTotalHomologado"] \
                         .sum().sort_values(ascending=True)

    plt.figure(figsize=(12, 6))
    colors = sns.color_palette("Set2", n_colors=len(valores_natureza))
    ax = valores_natureza.plot(kind="barh", color=colors)
    ax.xaxis.set_major_formatter(mtick.FuncFormatter(formatar_moeda))
    plt.title("Valores Homologados por Natureza Jurídica")
    plt.xlabel("Valor Total Homologado (R$)")
    plt.ylabel("Natureza Jurídica")
    add_values_bars(ax)
    plt.tight_layout()
    plt.subplots_adjust(left=0.3)  
    plt.show()
```

Soma de valorTotalHomologado agrupada por naturezaJuridicaNome. Mostra distribuição do gasto por tipo de entidade (ex.: Sociedade Anônima, Empresa Privada, Município, Cooperativa, etc.). 

---

# __3.3.Distribuição de Valores Homologados para MPEs__
Gráfico filtrando e mostrando valores homologados apenas para MPEs

```python
# --- Itens homologados para MPEs ---
if "aplicacaoBeneficioMeepp" in df.columns:
    itens_mpe = df[df["aplicacaoBeneficioMeepp"] == True]
    if not itens_mpe.empty:
        plt.figure(figsize=(12, 6))
        ax = sns.histplot(
            itens_mpe["valorTotalHomologado"],
            bins=15,
            stat="count",
            kde=True,
            color="orange"
        )
        plt.title("Distribuição de Valores Homologados para MPEs")
        plt.xlabel("Valor Total Homologado (R$)")
        plt.ylabel("Número de Itens")
        plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(formatar_moeda))
        plt.tight_layout()
        plt.subplots_adjust(left=0.3)  
        plt.show()
```

Filtragem por aplicacaoBeneficioMeepp == True e plotagem da distribuição dos valorTotalHomologado. Mostra a distribuição (concentração) dos valores homologados quando aplicou-se benefício a micro/pequenas empresas (MPEs).

---

# __3.4.Evolução do Valor Homologado ao Longo do Tempo__
Gráfico que mostra evolução do valor homologado ao longo do tempo

```python
# --- Evolução ao longo do tempo ---
if "ano_mes" in df.columns:
    evolucao_valores = df.groupby("ano_mes")["valorTotalHomologado"].sum().reset_index()
    if not evolucao_valores.empty:
        plt.figure(figsize=(12, 6))
        ax = sns.lineplot(data=evolucao_valores, x="ano_mes", y="valorTotalHomologado",
                          marker="o", color="blue")
        ax.yaxis.set_major_formatter(mtick.FuncFormatter(formatar_moeda))
        plt.title("Evolução do Valor Total Homologado ao Longo do Tempo")
        plt.xlabel("Ano-Mês")
        plt.ylabel("Valor Total Homologado (R$)")
        plt.xticks(rotation=45)
        add_values_line(ax, evolucao_valores["ano_mes"], evolucao_valores["valorTotalHomologado"])
        plt.tight_layout()
        plt.subplots_adjust(left=0.3)  
        plt.show()
```

Agregação por ano_mes (extraído de dataResultadoPncp) somando valorTotalHomologado. Permite identificar tendências temporais: aumentos, quedas e sazonalidade no volume financeiro homologado.

---

# __3.5.País de Origem dos Produtos e Serviços__
Gráfico mostrando o país de Origem de produtos e Serviços

```python
# --- Países de origem ---
if "codigoPais" in df.columns:
    paises = df["codigoPais"].value_counts()
    if not paises.empty:
        plt.figure(figsize=(12, 6))
        colors = sns.color_palette("Paired", n_colors=len(paises.head(10)))
        ax = paises.head(10).plot(kind="bar", color=colors)
        plt.title("Países de Origem dos Produtos/Serviços")
        plt.xlabel("País")
        plt.ylabel("Quantidade de Itens")
        for i, v in enumerate(paises.head(10)):
            ax.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10)
        plt.tight_layout()
        plt.subplots_adjust(left=0.3)
        plt.show()
```

Contagem por codigoPais (ou paisOrigemProdutoServicoId) e plot dos principais países. 

---
